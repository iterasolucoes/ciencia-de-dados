```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", palette="muted")
```


https://seaborn.pydata.org/

```python
df = pd.read_csv('../base/train.csv')
df.head()
```

```python
# Construct iris plot
sns.swarmplot(x="Pclass", y="Fare", data=df)

# Show plot
plt.show()
```

```python
g = sns.catplot("Pclass", "Survived", "Sex", data=df, kind="bar", palette="pastel")
                   
# Show plot
plt.show()
```

```python
sns.barplot(x="Sex", y="Survived", data=df)
```

```python
grid = sns.FacetGrid(df, col='Survived', row='Pclass', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age')
grid.add_legend();
```

```python
grid = sns.FacetGrid(df, row='Embarked', size=3, aspect=1.6)
grid.map(sns.pointplot, 'Pclass', 'Survived', 'Sex')
grid.add_legend()
```

```python
sns.lineplot(data=df[['Fare']], palette="tab10")
```

```python
sns.boxplot(x="Pclass", y="Age", palette=["m", "g"],
            data=df)
```

```python
df['Pclass'].value_counts().plot(kind='barh')
```
